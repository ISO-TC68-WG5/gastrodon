```python
endpoint=RemoteEndpoint(
    "https://semantics.technical-fixprotocol.org/fuseki/demo/sparql",
    prefixes=prefixes.graph,
    base_uri="urn:iso:std:iso:20022:2013:owl:metamodel#"
)
```

```python
predicates=endpoint.select("""
    SELECT ?p (COUNT(*) AS ?count) { ?s ?p ?o .} GROUP BY ?p ORDER BY DESC(?count)
""")
predicates
```
